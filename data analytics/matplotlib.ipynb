```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

## Functional Method

```python
x = np.linspace(0, 5, 11)
x
```

```python
y = x**2
y
```

```python
plt.plot(x, y)
#plt.show()
```

```python
plt.plot(x, y, 'r')
```

```python
plt.plot(x, y, '#8F00FF')
```

```python
plt.plot(x, y, 'g')
plt.title('My Graph')
plt.xlabel('X Array')
plt.ylabel('Y Array')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'y*-')

plt.tight_layout()
```

```python
plt.subplot(2, 2, 1)
plt.subplot(2, 2, 2)
plt.subplot(2, 2, 3)
plt.subplot(2, 2, 4)

plt.tight_layout()
```

## Object Oriented Method

```python
x 
```

```python
y
```

```python
canvas = plt.figure()

#add_axes([left margin, bottom margin, width, height]) between 0 and 1
axes = canvas.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
```

```python
#Creates a blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes1.plot(x, y)

axes2 = fig.add_axes([0.65, 0.2, 0.2, 0.2])
axes2.plot(y, x, 'g')
```

```python
newCanvas = plt.figure(figsize=(10,6))

axes = newCanvas.add_axes([0, 0, 1, 1])
axes.plot(x, y)
```
