## Text Filtering

```python
import nltk
```

```python
#nltk.download()
```

```python
from nltk.corpus import stopwords
```

```python
stopwords.words()
```

```python
stopwords.words('english')
```

```python
sampleText = "Good morning! How are you? I'm great."
```

```python
import string

string.punctuation
```

```python
noPunc = []

for char in sampleText:
    if char not in string.punctuation:
        noPunc.append(char)
```

```python
noPunc
```

```python
noPunc = ''.join(noPunc)
noPunc
```

```python
cleanMessage = []

for word in noPunc.split():
    if word not in stopwords.words('english'):
        cleanMessage.append(word)
        
cleanMessage
```

**Tokenization**

```python
def textFiltering(message):
    noPunc = [char for char in message if char not in string.punctuation]
    noPunc = ''.join(noPunc)
    noPunc = [word for word in noPunc.split() if word.lower() not in stopwords.words('english')]
    return noPunc
```

```python
df.head()
```

```python
df['message'].head().apply(textFiltering)
```

## Count Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
```

**Bag Of Words**

```python
bagOfWords = CountVectorizer(analyzer=textFiltering).fit(df['message'])
bagOfWords
```

```python
bagOfWords.vocabulary_
```

```python
len(bagOfWords.vocabulary_)
```

```python
sampleMessage = df['message'][3]
sampleMessage
```

```python
sampleBag = bagOfWords.transform([sampleMessage])
print(sampleBag)
```

```python
messageBow = bagOfWords.transform(df['message'])
```

```python
print(messageBow)
```

```python
messageBow.shape
```

## TF-IDF Transformer

TF: Term Frequency <br>
IDF: Inverse Document Frequency

```python
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
tfidf = TfidfTransformer().fit(messageBow)
tfidf
```

```python
sampleTfidf = tfidf.transform(sampleBag)
print(sampleTfidf)
```

```python
messagesTfidf = tfidf.transform(messageBow)
print(messagesTfidf)
```

## Naive Bayes Classifier

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB().fit(messagesTfidf, df['label'])
model
```

```python
model.predict(sampleTfidf)
```
