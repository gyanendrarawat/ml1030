# Exploratory Data Analysis (EDA)

## Distribution Plots

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.shape
```

```python
tips['day'].unique()
```

```python
tips['time'].unique()
```

```python
tips['smoker'].unique()
```

```python
tips['total_bill'].unique()
```

```python
tips.info()
```

```python
tips['size'].unique()
```

```python
sns.set_style('darkgrid')
```

```python
plt.figure(figsize=(10,8))
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], kde=False)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex', cmap='rocket')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex', palette='husl')
```

## Categorical Plots

```python
tips.head()
```

```python
sns.countplot(x='sex', data=tips, palette='husl')
```

```python
plt.figure(figsize=(12,10))
sns.boxplot(x='day', y='total_bill', data=tips, palette='rainbow')
```

```python
plt.figure(figsize=(12,10))
sns.boxplot(x='day', y='total_bill', data=tips, palette='rainbow', hue='smoker')
```

## Matrix Plots

```python
tips.corr()
```

```python
sns.heatmap(tips.corr())
```

```python
sns.heatmap(tips.corr(), annot=True)
```

## Regression Plots

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='husl')
```
